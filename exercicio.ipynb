```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('dados.csv', sep=';')
df
```

```python
sns.histplot(data = df['PIB'], kde = True, bins = 10, color='blue')
```

```python
sns.histplot(data = df['VALOREMPENHO'], kde = True, bins = 10, color='red')
```

```python
municipios_top_10 = df.sort_values(by='PIB', ascending=False).head(10)['MUNICIPIO']
sns.barplot(data=df, x=municipios_top_10, y='PIB', color = 'k',estimator = sum)
plt.xticks(rotation=90)
plt.show()
```

```python
municipios_top_10 = df.sort_values(by='PIB', ascending=False).head(10)['MUNICIPIO']
sns.barplot(data=df, x = municipios_top_10, y='VALOREMPENHO', color = 'k',estimator = sum)
plt.xticks(rotation=90)
plt.show()
```
